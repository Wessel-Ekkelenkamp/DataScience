```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("klassenpresentatie.xlsx",
                    header=None,
                    skiprows=2, 
                    usecols=range(1,10),
                    )
df.columns = df.iloc[0]
df = df[1:]

df['liever op school of thuis'] = df['liever op school of thuis'].str.upper()

df['liever op school of thuis'].value_counts().plot.bar()

plt.show()



```
